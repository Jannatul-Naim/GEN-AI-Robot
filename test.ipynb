```python
from brain.agent import BrainAgent
from brain.llm_engine import LLMEngine

yolo_data = {
  "objects": [
    {
      "name": "cup",
      "confidence": 0.74,
      "distance_cm": 23.8,
      "center": [623, 585],
      "grasp_center": [623, 605]
    },
    {
      "name": "bottle",
      "confidence": 0.69,
      "distance_cm": 30.4,
      "center": [471, 508],
      "grasp_center": [471, 528]
    }
  ]
}

brain = BrainAgent(LLMEngine())

print(brain.process("pick bottle", yolo_data))
print(brain.process("pick cup", yolo_data))
print(brain.process("stop", yolo_data))
print(brain.process("what is around me?", yolo_data))
print(brain.process("pick book", yolo_data))  
```
